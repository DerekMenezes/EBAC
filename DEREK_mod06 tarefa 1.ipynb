```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nasc_sus = pd.read_csv('sinasc_RO_2019.csv')
```

```python
nasc_sus.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(20, 20))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0,0], x='APGAR1',data=nasc_sus)
sns.countplot(ax = axes[0,1], x='APGAR5',data=nasc_sus)
sns.countplot(ax = axes[0,2], x='SEXO',data=nasc_sus)
sns.countplot(ax = axes[1,0], x='GESTACAO',data=nasc_sus).tick_params("x", rotation=45)
sns.countplot(ax = axes[1,1], x='ESTCIVMAE',data=nasc_sus)



plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

nasc_sus["APGCON"]= nasc_sus["APGAR5"]/nasc_sus["CONSULTAS"]
plt.rc('figure', figsize=(10, 10))

fig, ax = plt.subplots()

sns.countplot(x='APGCON',data=nasc_sus).tick_params("x",rotation=90)
#ax.bar(nasc_sus.index, nasc_sus["APGCON"])
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(20, 20))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0,0], x='SEXO',data=nasc_sus)
sns.countplot(ax = axes[0,1], x='GESTACAO',data=nasc_sus).tick_params("x", rotation=45)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE',data=nasc_sus)
sns.countplot(ax = axes[1,1], x='CONSULTAS',data=nasc_sus)
```
